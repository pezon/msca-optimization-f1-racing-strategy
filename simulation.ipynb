# Race Simulation

```julia
using TOML

include("src/race_times.jl")
include("src/strategies.jl")
include("src/simulation.jl")

params = TOML.parsefile("params/params-empirical-lap39.toml")
# params = TOML.parsefile("params/params-cbeaume.toml")
```

```julia
results = simulate(;
    tires=params["simulation"]["tires"],
    min_pit_stops=params["simulation"]["min_pit_stops"],
    max_pit_stops=params["simulation"]["max_pit_stops"],
    t_base=params["simulation"]["t_base"],
    tot_no_laps=params["simulation"]["tot_no_laps"],
    t_lap_sens_mass=params["simulation"]["t_lap_sens_mass"],
    t_pitdrive_inlap=params["simulation"]["t_pitdrive_inlap"],
    t_pitdrive_outlap=params["simulation"]["t_pitdrive_outlap"],
    t_pit_tirechange=params["simulation"]["t_pit_tirechange"],
    pits_aft_finishline=params["simulation"]["pits_aft_finishline"],
    cold_tire_pen=params["simulation"]["cold_tire_pen"],
    p_grid=params["simulation"]["p_grid"],
    t_loss_pergridpos=params["simulation"]["t_loss_pergridpos"],
    t_loss_firstlap=params["simulation"]["t_loss_firstlap"],
    m_fuel_init=params["simulation"]["m_fuel_init"],
    b_fuel_perlap=params["simulation"]["b_fuel_perlap"],
    tire_pars=params["tires"]
)
```

```julia
tires(rs) = [stint[3] for stint in rs]
laps(rs) = [(stint[1], stint[2]) for stint in rs]
results[:,"Tires"] = tires.(results[:,"Race_Strategy"])
results[:,"Laps"] = laps.(results[:,"Race_Strategy"])
unique(results,3)
```
